```python
# Import necessary libraries
import pandas as pd
import pandas_datareader as pdr
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Enable retrieved stock data to be stored in a pandas dataframe
yf.pdr_override()

# Download MAANG stock closing price data from Yahoo and store in a dataframe
stock_df = pdr.data.get_data_yahoo(["META", "AMZN", "AAPL", "NFLX", "GOOG"], start="2020-01-01", end="2023-12-31")["Close"]
stock_df
```

```python
# Write closing price data to csv for import into Tableau
stock_df.to_csv("MAANG_closing_price.csv")
```

```python
# Plot closing price data for each stock for initial inspection
plt.figure(figsize=(15,10))
plt.plot(stock_df)
plt.title('MAANG Close Price History', fontsize=18)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Stock Price ($)', fontsize=16)
plt.legend(["AAPL", "AMZN", "GOOG", "META", "NFLX"], loc='upper left')
```
